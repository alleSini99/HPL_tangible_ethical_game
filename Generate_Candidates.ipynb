```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
sample_size = 40
```

```python
#At EPFL 30.3% female students, 1 = female
gender = (np.random.rand(sample_size) < 0.303).astype(int)


#20.8% of the swiss population is disable but and 64% of this works Not sure what to do here.
disability = (np.random.rand(sample_size) < 0.208).astype(int)

# 29% of workers in stem are members of an ethnical minority (in USA, no data for EPFL or Switzerland)
minority = (np.random.rand(sample_size) < 0.29).astype(int)

# 9.22% is the official swiss statistic. Probably way less at EPFL.
criminal = (np.random.rand(sample_size) < 0.12).astype(int)

# ranking is from 0 to 200
mu, sigma = 65, 30 # mean and standard deviation
university_ranking = np.random.normal(mu, sigma, sample_size).astype(int)
university_ranking[university_ranking<1]=1

mu, sigma = 3, 5 # mean and standard deviation
work_experience = np.random.normal(mu, sigma, sample_size).astype(int)
work_experience[work_experience<0]=0

mu, sigma = 5, 1 # mean and standard deviation
studying = np.random.normal(mu, sigma, sample_size).astype(int)
studying[studying<0]=0

mu, sigma = 75, 15 # mean and standard deviation
GPA = abs(np.random.normal(mu, sigma, sample_size)).astype(int)
GPA[GPA>100]=100

mu, sigma = 75, 15 # mean and standard deviation
language = abs(np.random.normal(mu, sigma, sample_size)).astype(int)
language[language>100]=100

mu, sigma = 75, 15 # mean and standard deviation
com = abs(np.random.normal(mu, sigma, sample_size)).astype(int)
com[com>100]=100

age_structure = np.array([920, 1150, 1200, 960, 730, 500])
probabilities = np.zeros(60)
for i in range(age_structure.size):
    probabilities[i*10:(i+1)*10] = age_structure[i]/(10*np.sum(age_structure)) 

age = np.random.choice(np.arange(20, 80), size = sample_size, p = probabilities) 

mu, sigma = 1.53, 2 # mean and standard deviation
kids = abs(np.random.normal(mu, sigma, sample_size)).astype(int)
kids[kids<0]=0
```

```python
df = pd.DataFrame({'gender': gender, 'University Rank': university_ranking, 'years of studies':studying, 'GPA':GPA, 
                    'Language':language , 'Work experience': work_experience, 
                   'Communication skills':com, '# children':kids, 'Disability':disability,
                  'Ethnical minority': minority, 'Age': age, "Criminal records": criminal})
df



```
